# Handling missing values

```python
# there are missing values
data.isna().sum()
```

```python
def get_missing_val_count_df(df):
    missing_values = df.isnull().sum()
    missing_df = pd.DataFrame({'Feature_with_missing_value': missing_values.index, 'Missing_values': missing_values.values})
    return missing_df[missing_df['Missing_values'] > 0]
```

```python
msdf = get_missing_val_count_df(data)
msdf
```

```python
def process_columns(df, column_names):
    for column_name in column_names:
        for i in range(df.shape[0]):
            if df.loc[i, column_name] < 0.5:
                df.loc[i, column_name] = 0
            else:
                df.loc[i, column_name] = 1
```

```python
process_columns(binary_data, binary_cols)
```

```python
import math

fig, axes = plt.subplots(nrows=math.ceil(len(cat_data.columns) / 2), ncols=2, figsize=(10, 90))
fig.suptitle('Distributions of Categorical Features', fontsize=15)

for i, col in enumerate(cat_data.columns):
        row_index = i // 2  # Calculate the row index for the subplot
        col_index = i % 2   # Calculate the column index for the subplot
        sns.countplot(x=cat_data[col], ax=axes[row_index, col_index])
        axes[row_index, col_index].set_title(col, fontsize=10)
        axes[row_index, col_index].set_xlabel(None)
        axes[row_index, col_index].set_ylabel('count', fontsize=10)

# If there are an odd number of columns, remove the empty subplot
if len(cat_data.columns) % 2 != 0:
    fig.delaxes(axes[math.ceil(len(cat_data.columns) / 2) - 1, 1])

plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust layout to prevent overlap
plt.show()
```

```python
health_condition = pd.get_dummies(category_data['Health_cond'], drop_first=True,dtype=int)
category_data.drop(['Health_cond'], axis=1, inplace=True)
category_data = pd.concat([category_data, health_condition], axis =1)
category_data = category_data.rename(columns={1:'HC1', 2: 'HC2', 3: 'HC3', 4:'HC4',5:'HC5'}) # 4 binary columns enough to describe 5 categories
```

```python
days_hospitalised = pd.get_dummies(category_data['days_hospitalised'], drop_first=True,dtype=int)
category_data.drop(['days_hospitalised'], axis=1, inplace=True)
category_data = pd.concat([category_data, days_hospitalised], axis =1)
category_data = category_data.rename(columns={1:'days_hospitalised_1', 2: 'days_hospitalised_2', 3: 'days_hospitalised_3', 4:'days_hospitalised_4',5:'days_hospitalised_5',6:'days_hospitalised_6'})
category_data.head()
```
