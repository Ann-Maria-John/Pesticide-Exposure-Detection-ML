```python
identity= pd.read_csv("C:/2nd/Machine Learning/Project/Datasets/age-gender.csv")
identity=identity.drop(identity.columns[0], axis=1)

# Define a function to classify individuals
def identification(row):
    if row['DMDMARTL'] == 1:
        return "married"
    if row['RIDEXPRG'] == 1:
        return "pregnant"
    
# Apply the classification function to create two new target variables
# 1 =Male ,2=Female
identity['Gender']=identity['RIAGENDR']
identity['Age']=identity['RIDAGEYR']
identity['Marital_status'] = identity.apply(lambda row: 1 if identification(row) == "married" else 0, axis=1)
identity['Pregnant'] = identity.apply(lambda row: 1 if identification(row) == "pregnant" else 0, axis=1)
identity=identity.iloc[:, :1].join(identity.iloc[:, -4:])
identity
```

##### Alcohol Conversion


```python
alc_df= pd.read_csv("C:/2nd/Machine Learning/Project/Datasets/alcohol use.csv")
alc_df=alc_df.drop(alc_df.columns[0], axis=1)
# Define a function to classify individuals

def classify_alcohol_consumption(row):
    if row['ALQ101'] == 1:
        return "current_drinker"
    elif row['ALQ110'] == 1:
        return "past_drinker"
    
# Apply the classification function to create two new target variables
alc_df['current_drinker'] = alc_df.apply(lambda row: 1 if classify_alcohol_consumption(row) == "current_drinker" else 0, axis=1)
alc_df['former_drinker'] = alc_df.apply(lambda row: 1 if classify_alcohol_consumption(row) == "past_drinker" else 0, axis=1)
alcohol=alc_df.iloc[:, :1].join(alc_df.iloc[:, -2:])
alcohol.shape
```

##### Smoker

```python
smoke_df= pd.read_csv("C:/2nd/Machine Learning/Project/Datasets/Cigaratte use.csv")
smoke_df=smoke_df.drop(smoke_df.columns[0], axis=1)

def classify_smoker(row):
    if row['SMQ020'] == 1 and row['SMQ040'] <= 2:
        return "current_smoker"
    elif row['SMQ020'] == 1 and row['SMQ040'] == 3:
        return "former_smoker"
    
# Apply the classification function to create two new target variables
smoke_df['current_smoker'] = smoke_df.apply(lambda row: 1 if classify_smoker(row) == "current_smoker" else 0, axis=1)
smoke_df['former_drinker'] = smoke_df.apply(lambda row: 1 if classify_smoker(row) == "former_smoker" else 0, axis=1)
smoker=smoke_df.iloc[:, :1].join(smoke_df.iloc[:, -2:])
smoker.shape
```

##### Merging dataframes

```python
merged_data=pd.merge(identity,alcohol,on='SEQN',how='outer')
merged_data=pd.merge(merged_data,smoker,on='SEQN',how='outer')
merged_data=pd.merge(merged_data,Aspirin_use,on='SEQN',how='outer')
merged_data=pd.merge(merged_data,BP_Chol,on='SEQN',how='outer')
merged_data=pd.merge(merged_data,chest_pain,on='SEQN',how='outer')
merged_data=pd.merge(merged_data,stomach,on='SEQN',how='outer')
merged_data=pd.merge(merged_data,Diabetes,on='SEQN',how='outer')
merged_data=pd.merge(merged_data,Disability,on='SEQN',how='outer')
merged_data=pd.merge(merged_data,Drug_addict,on='SEQN',how='outer')
merged_data=pd.merge(merged_data,liver,on='SEQN',how='outer')
merged_data=pd.merge(merged_data,hospital,on='SEQN',how='outer')
merged_data=pd.merge(merged_data,kidney,on='SEQN',how='outer')
merged_data=pd.merge(merged_data,pesticide_use,on='SEQN',how='outer')
merged_data=pd.merge(merged_data,medical_cond,on='SEQN',how='outer')
merged_data=pd.merge(merged_data,occupation_type,on='SEQN',how='outer')
merged_data=pd.merge(merged_data,childhood,on='SEQN',how='outer')
merged_data=pd.merge(merged_data,medicine_use,on='SEQN',how='outer')
merged_data
```
